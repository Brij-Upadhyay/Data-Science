```python
#pip install kafka-python
from kafka import KafkaProducer
import json
import pandas as pd
import random
import numpy as np
import time
```

```python
pizza_center_loc=np.array([20,30])

move_li=['+','-']
n=100


df=pd.read_csv(r"C:\Users\briju\Desktop\Mtech\Assignment\Data\pizza_customers.csv",dtype=str)
producer=KafkaProducer(bootstrap_servers={'localhost:9092'},api_version=(0,10,1))

def dist(p1,p2):
    squared_dist = np.sum((p1-p2)**2, axis=0)
    dist = np.sqrt(squared_dist)
    return dist
```

```python
def person_loc_generator(a): 
    initial_loc=np.array([0,0])
    prev_loc=np.array([1,1])
    status_n=0
    status_f=0
    for i in range(n):
        val=random.randint(1,4)
        prev_loc=np.ndarray.copy(initial_loc)
        move_x=move_li[random.randint(0,1)]
        move_y=move_li[random.randint(0,1)]
        print(move_x,move_y)
        if move_x=='+' and move_y=='-':
            initial_loc[0]=initial_loc[0]+val
            initial_loc[1]=initial_loc[1]-val
        elif move_x=='+' and move_y=='+':
            initial_loc[0]=initial_loc[0]+val
            initial_loc[1]=initial_loc[1]+val
        elif move_x=='-' and move_y=='+':
            initial_loc[0]=initial_loc[0]-val
            initial_loc[1]=initial_loc[1]+val
        elif move_x=='-' and move_y=='-':
            initial_loc[0]=initial_loc[0]-val
            initial_loc[1]=initial_loc[1]-val
        #print(prev_loc,initial_loc)
        prev_loc_dist=dist(prev_loc,pizza_center_loc)
        new_loc_dist=dist(initial_loc,pizza_center_loc)
        print(prev_loc_dist,new_loc_dist)
        if prev_loc_dist>new_loc_dist and new_loc_dist<50:
            status_n=status_n+1
            if status_n<2:
                print("offer can be generatedas person is coming near ")
                b={'id':str(a),'dir':'N'}
                producer.send("test-topic",json.dumps(b).encode('utf-8'))

        if prev_loc_dist<new_loc_dist and new_loc_dist>50:
            status_f=status_f+1
            if status_f<2 and status_n>1:
                b={'id':str(a),'dir':'F'}
                producer.send("test-topic",json.dumps(b).encode('utf-8'))

                print("bigger offer can be generated  as person is going away")
        time.sleep(0.5)
        if status_f>1 and status_n>1:
            break
```

```python
#below code will generate offer if person will come near pizza store and if its going away from pizza
```

```python
for i in range(10):
    a=df.at[random.randint(0,200),"CustomerID"]
    person_loc_generator(a)
```
